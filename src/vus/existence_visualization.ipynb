# Visual explanation of existence

## FF-VUS (CPU version)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from vus_numpy import VUSNumpy
```

```python
# Create toy examples
label = np.array([0] * 100 + [1] * 20 + [0] * 100)
score = np.random.uniform(0, 0.05, label.shape[0])
score[110] = 0.7
```

```python
fig, axs = plt.subplots(2, 1, figsize=(8, 4), sharex=True)

axs[0].plot(label, markersize=3)
axs[0].set_title('Label')
axs[0].set_ylabel('Label')

axs[1].plot(score, markersize=3)
axs[1].set_title('Score')
axs[1].set_ylabel('Score')
axs[1].set_xlabel('Index')
axs[1].set_ylim(0, 1)

plt.tight_layout()
plt.show()
```

```python
ffvus = VUSNumpy(10)
```

```python
# Return the time_it as it was and implement this in the vus visualizer file

(start_points, end_points), _ = ffvus.get_anomalies_coordinates(label)
slopes = ffvus.add_slopes_precomputed(label, start_points, end_points)

sns.lineplot(slopes.T)
plt.show()
```
